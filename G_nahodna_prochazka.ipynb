## Náhodná procházka

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
N = 100 # delka nahodneho vektoru X 
A = np.zeros((N, N))
for i in range(len(A)):
    A[i, 0:i+1] = np.ones(i+1)
S = np.dot(A, A.T) # kovariancni matice
```

```python
R = np.random.normal(size=N)
x = np.dot(A, R) # vzorek nahodneho vektoru s rozdelenim N(0, S)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x, 'r', label=('$X_t$'))
plt.plot(2*S.diagonal()**0.5, 'k', label='$\mu_t\pm 2\sigma_t$')
plt.plot(-2*S.diagonal()**0.5, 'k')
plt.xlabel('$t$')
plt.ylabel('$X_t$')
plt.legend()
```
